```python
###### import json
from datetime import datetime


def input_student(subjects):
    name= input("enter student name: ").strip()
    marks = {}
    for sub in subjects:
        while True:
            try:
                val = float(input(f"enter marks in {sub} (0-100):"))
                if 0 <= val <=100:
                    marks[sub] = val
                    break
                else:
                    print("marks must be between 0 and 100.")
            except valueerror:
                print("invalid input, enter a number.")
    return {"name": name, "marks": marks}
def evaluate_student(student, subjects):
    total = sum(student["marks"][s] for s in subjects)
    percentage = total / len(subjects)
    if percentage >= 90:
        grade = "A+"
    elif percentage >= 80:
        grade = "A"
    elif percentage >= 70:
        grade = "B"
    elif percentage >= 60:
        grade = "C"
    elif percentage >= 50:
        grade = "D"
    else:
        grade = "F"
    student["total"] = round(total, 2)
    student["percentage"] = round(percentage, 2)
    student["grade"] = grade
    return student
def enforce_min_pass(student, min_mark=33):
    if any(m < min_mark for m in student["marks"].values()):
        student["grade"] = "F"
def print_report(student, subjects):
    print("-" * 40)
    print(f"report card: {student['name']}")
    print("-" * 40)
    for s in subjects:
        print(f"{s:<12}: {student['marks'][s]:>6.2f}")
    print("-" * 40)
    print(f"{'total' :<12}: {student['total']:>6.2f}")
    print(f"{'percent' :<12}: {student['percentage']:>6.2f}%")
    print(f"{'grade':<12}: {student['grade']}")
    print("-" * 40)
def save_to_file(students, filename="report_card.json"):
    with open(filename,"w") as f:
        json.dump(students,f,indent=2)
def load_from_file(filename="report_cards.json"): 
    try:
        with open(filename, "r") as f:
            return json.load(f)
    except FileNotFoundError:
        return []
def print_all(students, subjects):
    if not students:
        print("no record founds.")
        return
    for s in students:
        print_report(s, subjects)
def search_student(students, name):
    name = name.strip().lower()
    return [s for s in students if s ["name"].lower() == name]
def log_event(message, filename="report_log.txt"):
    ts = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open(filename, "a") as f:
        f.write(f"[{ts}] {message}\n")
def main():
    subjects = ["math", "physics", "chemistry", "english", "cs",]
    students = load_from_file()
    while True:
        print("\nmenu:")
        print("1. add student")
        print("2. show all reports")
        print("3. search by name")
        print("4. save")
        print("5. load")
        print("6. exit")
        choice = input ("choose (1-6): ").strip()

        if choice == "1":
            s = input_student(subjects)
            s = evaluate_student(s, subjects)
            enforce_min_pass(s)
            students.append(s)
            log_event(f"added student: {s['name']}")
            print("student added.")
        elif choice == "2":
            print_all(students, subjects)
        elif choice == "3":
            q = input("enter name to search: ")
            results = search_student(students, q)
            if results:
                for r in results:
                    print_report(r, subjects)
            else:
                print("no match found.")
        elif choice == "4":
            save_to_file(students)
            print("saved to report_cards.json")
        elif choice == "5":
            students = load_from_file()
            print(f"loaded {len(students)} records.")
        elif choice == "6":
            print("goodbye!")
            break
        else:
            print("invalid choice.")

if __name__ == "__main__":
    main()
```
